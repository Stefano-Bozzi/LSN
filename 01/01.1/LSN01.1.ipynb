```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("output1.01.1.txt", usecols=(0,1,2), delimiter='\t', unpack='true')
plt.errorbar(x,f,error)
plt.title('Mean estimation')
plt.xlabel('#bloks')
plt.ylabel('<r>')
plt.grid(True)
plt.plot([0,100],[0.5,0.5], color="red")
plt.show()
```
