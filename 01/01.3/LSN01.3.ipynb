```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, f, error = np.loadtxt("pi.txt", usecols=(0,1,2), delimiter='\t', unpack='true')
plt.errorbar(x,f,error)
plt.title('Mean estimation of pi')
plt.xlabel('#throws')
plt.ylabel('<pi>')
plt.grid(True)
plt.plot([0,100],[math.pi,math.pi], color="red")
plt.show()
```
